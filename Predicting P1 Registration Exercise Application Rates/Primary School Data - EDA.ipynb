```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('primary_school_clean.csv')  
```

```python
df1 = df[(df['Year'] >= 2019) & (df['Phase']=='1')]
```

```python
# most popular primary schools based on application rate
df1_sorted = df1.sort_values(['Year', 'Application Rate'], ascending=False).groupby('Year').head(10)
df1_sorted.reset_index(inplace=True)
df1_sorted
```

```python
df1_sorted['School'].value_counts()
```

```python
df1_2022 = df1_sorted[df1_sorted.Year==2022].sort_values(['Year', 'Application Rate'], ascending=False)
df1_2022
```

```python
df_2022 = df1_sorted[df1_sorted.Year==2022].sort_values(['Year', 'Application Rate'], ascending=False)
ax = sns.barplot(x='Application Rate', y='School', data=df1_2022, orient='h')
ax.set_xlabel('Application Rate')
ax.set_ylabel('Schools')    

plt.title('Top 10 Most Popular Schools in Phase 1 (2022)')
# ax.bar_label(ax.containers[0], fmt= '%0.1f' )
plt.show()
```

Phase 1 is for a child who has a sibling studying in the primary school.
Popular schools in Phase 1 are all mixed genders schools and all have more than half of vacancies at Phase 1 filled.

```python
df2A = df[(df['Year'] >= 2019) & (df['Phase']=='2A')]
```

```python
# most popular primary schools based on application rate
df2A_sorted = df2A.sort_values(['Year', 'Application Rate'], ascending=False).groupby('Year').head(10)
df2A_sorted
```

```python
df2A_sorted['School'].value_counts()
```

Phase 2A is for a child:
Whose parent or sibling is a former student of the primary school, including those who have joined the alumni association of the primary school as a member.
Whose parent is a member of the School Advisory or Management Committee.
Whose parent is a staff member of the primary school. 
From the MOE Kindergarten under the purview of and located within the primary school. 

We can infer that popular schools in Phase 2A are schools with traditionally strong alumni ties like Nan Hua, Pei Hwa Presbyterian, Rosyth, CHIJ St. Nicholas Girls, Catholic High, Nanyang etc. These schools also tend to be either GEP or SAP schools and parents probably feel that such schools have a better reputation. 

Interestingly, schools like Gongshang, Westwood, Punggol Green have also became very popular in the recent years at Phase 2A. It could be due to the presence of MOE kindergarten within the school.

```python
df2B = df[(df['Year'] >= 2019) & (df['Phase']=='2B')]
```

```python
# most popular primary schools based on application rate
df2B_sorted = df2B.sort_values(['Year', 'Application Rate'], ascending=False).groupby('Year').head(10)
df2B_sorted
```

```python
df2B_sorted['School'].value_counts()
```

Phase 2B is for a child:
Whose parent has joined the primary school as a parent volunteer not later than 1 July of the year before P1 registration and has given at least 40 hours of voluntary service to the school by 30 June of the year of P1 registration.
Whose parent is a member endorsed by the church or clan directly connected with the primary school.
Whose parent is endorsed as an active community leader.

Balloting is very intense under Phase 2B where it does not even confirm a slot if you live within 1km of the school of your choice. We see that the schools which are popular in Phase 2B are affiliated to churches like CHIJ St. Nicholas Girls, Catholic High, Methodist Girls, St. Josephs Institution Junior etc or schools affiliated to clans like Tao Nan, Ai Tong, Chongfu etc

```python
df2C = df[(df['Year'] >= 2019) & (df['Phase']=='2C')]
```

```python
# most popular primary schools based on application rate
df2C_sorted = df2C.sort_values(['Year', 'Application Rate'], ascending=False).groupby('Year').head(10)
df2C_sorted
```

```python
df2C_sorted['School'].value_counts()
```

Phase 2C is for a child who is ineligible for or unsuccessful in earlier phases. It is usually distance based. Balloting is even more intense in Phase 2C with as many as 5-6 people fighting over 1 vacancy. It is certain that balloting happens for these schools within 1km so it is very risky to bet on these schools by moving within 1km of the school. Schools like Nan Hua, Northland, South View have been consistently on this top 10 list of the past 4 years.
